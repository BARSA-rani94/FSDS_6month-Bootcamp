```python
x = df.drop('Loan_Status',axis = 1)
y = df['Loan_Status']
```

```python
# Ensure necessary imports are at top of this cell
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

Train = []
Test = []
CV = []

for i in range(0, 101):
    # split first (avoid data leakage)
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=i)

    # impute missing values using most frequent (suitable for Self_Employed categorical)
    imputer = SimpleImputer(strategy='most_frequent')
    x_train = pd.DataFrame(imputer.fit_transform(x_train), columns=x_train.columns, index=x_train.index)
    x_test = pd.DataFrame(imputer.transform(x_test), columns=x_test.columns, index=x_test.index)

    # train logistic regression
    log_default = LogisticRegression()
    log_default.fit(x_train, y_train)

    ypred_train = log_default.predict(x_train)
    ypred_test = log_default.predict(x_test)

    Train.append(accuracy_score(y_train, ypred_train))
    Test.append(accuracy_score(y_test, ypred_test))

    cv_scores = cross_val_score(log_default, x_train, y_train, cv=5, scoring='accuracy')
    CV.append(cv_scores.mean())

em = pd.DataFrame({'Train': Train, 'Test': Test, 'CV': CV})

# fix boolean arithmetic: compute absolute differences first, then compare
gm = em[(abs(em['Train'] - em['Test']) <= 0.05) & (abs(em['Test'] - em['CV']) <= 0.05)]

# handle case where gm might be empty
if not gm.empty:
    rs = gm[gm['CV'] == gm['CV'].max()].index.to_list()[0]
    print('best random_state number:', rs)
else:
    print("No suitable random_state found (no rows satisfied the tolerance criteria).")
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = rs)
```

# Machine Learning Modelling & Evaluation

# 1. Logostic Regression

```python
from sklearn.pipeline import make_pipeline

# use a pipeline that imputes missing values (most frequent) before fitting LogisticRegression
log_model = make_pipeline(SimpleImputer(strategy='most_frequent'), LogisticRegression())
log_model.fit(x_train, y_train)

ypred_train = log_model.predict(x_train)
ypred_test = log_model.predict(x_test)

print('Train Accuracy:', accuracy_score(y_train, ypred_train))
print('Cross Validation Score:', cross_val_score(log_model, x_train, y_train, cv=5, scoring='accuracy'))
print('Test Accuracy:', accuracy_score(y_test, ypred_test))
```

# 2.KNN

```python
from sklearn.neighbors import KNeighborsClassifier

# Use a pipeline that imputes missing values before fitting KNN to avoid NaNs error.
# SimpleImputer and make_pipeline are already imported in earlier cells, so reuse them.
estimator = make_pipeline(SimpleImputer(strategy='most_frequent'), KNeighborsClassifier())

# when using a pipeline, target the KNeighborsClassifier step name for hyperparameters
param_grid = {'kneighborsclassifier__n_neighbors': list(range(1, 50))}

from sklearn.model_selection import GridSearchCV
knn_grid = GridSearchCV(estimator, param_grid, scoring='accuracy', cv=5, n_jobs=-1)
# use the existing x_train/x_test variables (not X_train/X_test)
knn_grid.fit(x_train, y_train)

knn_model = knn_grid.best_estimator_

ypred_train = knn_model.predict(x_train)
ypred_test = knn_model.predict(x_test)

print('Train Accuracy:', accuracy_score(y_train, ypred_train))
print('Cross validation score:', cross_val_score(knn_model, x_train, y_train, cv=5, scoring='accuracy'))
print('Test Accuracy:', accuracy_score(y_test, ypred_test))
```
