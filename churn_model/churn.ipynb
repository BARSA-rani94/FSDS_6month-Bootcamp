```python
import  numpy as np
import pandas as pd
import tensorflow as tf
```

```python
tf.__version__
```

```python
dataset = pd.read_csv(r"C:\Users\HP\Downloads\Churn_Modelling.csv")
dataset.head()
```

```python
# select feature columns (CreditScore .. EstimatedSalary) and target column (Exited)
X = dataset.iloc[:,3:-1].values
y = dataset.iloc[:,-1].values
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X[:,2]=le.fit_transform(X[:,2])
```

```python
X
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1])],remainder='passthrough')
X=np.array(ct.fit_transform(X))
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# ANN model building Architechture

```python
ann=tf.keras.models.Sequential()  # Establish layer one after another
```

```python
ann.add(tf.keras.layers.Dense(units=6)) # Input layer
```

```python
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
ann.add(tf.keras.layers.Dense(units = 6,activation = 'relu'))
ann.add(tf.keras.layers.Dense(units = 5,activation = 'relu'))
ann.add(tf.keras.layers.Dense(units = 4,activation = 'relu')) # Hidden layer1
```

```python
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid')) # Output layer2
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train,batch_size = 32,epochs = 150,validation_data=(X_test,y_test))
```

```python
y_pred=ann.predict(X_test)
```

```python
y_pred=(y_pred>0.5)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),axis=1))  
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

ac = accuracy_score(y_test,y_pred)
ac
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
