```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import load_img, img_to_array
 

datagen=ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.4,
    height_shift_range=0.4,
    shear_range=0.4,
    zoom_range=0.4,
    horizontal_flip=True,
    rescale=None,
    fill_mode='constant'
)

img=load_img(r"C:\Users\HP\OneDrive\Desktop\training\Happy_faces\5.jpeg")
```

```python
img
```

```python
x=img_to_array(img)
x=x.reshape((1,)+x.shape)
i=0
for batch in datagen.flow(x,batch_size=1,save_to_dir=r"C:\Users\HP\OneDrive\Desktop\training\Happy_faces\happy",
                          save_prefix='happy',save_format='jpeg'):
    i+=1
    if i>30:
        break
```
