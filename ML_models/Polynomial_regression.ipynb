```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"C:\Users\HP\Downloads\New folder\emp_sal.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,1:2].values
y=dataset.iloc[:,2].values
```

```python
x
```

```python
y
```

```python
#linear model--linear algo(degree-1)
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
# polynomial model  ( bydefeaut degree - 2)

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=5)
x_poly = poly_reg.fit_transform(x)

poly_reg.fit(x_poly, y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)
```

```python
# linear regression visualizaton 
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg.predict(x), color = 'blue')
plt.title('Linear Regression graph')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# poly nomial visualization 

plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg_2.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# predicton 

lin_model_pred = lin_reg.predict([[6.5]])
lin_model_pred

poly_model_pred = lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
poly_model_pred
```

```python
#SVR model 
from sklearn.svm import SVR
svr_model=SVR(kernel='poly',degree=4,gamma='auto',C=10.0)
svr_model.fit(x,y)
```

```python
svr_model_pred=svr_model.predict([[6.5]])
print(svr_model_pred)
```

```python
#knn regressor model
from sklearn.neighbors import KNeighborsRegressor
knn_model=KNeighborsRegressor()
knn_model.fit(x,y)
```

```python
knn_model_pred=knn_model.predict([[6.5]])
print(knn_model_pred)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn_model=KNeighborsRegressor(n_neighbors=5,weights="distance",algorithm='brute',p=1)
knn_model.fit(x,y)
```

```python
knn_model_pred=knn_model.predict([[6.5]])
print(knn_model_pred)
```

```python
#Decision tree model
from sklearn.tree import DecisionTreeRegressor
dt_model=DecisionTreeRegressor()
dt_model.fit(x,y)
```

```python
dt_model_pred=dt_model.predict([[6.5]])
print(dt_model_pred)
```

```python
#Random forest
from sklearn.ensemble import RandomForestRegressor
rf_model=RandomForestRegressor()
rf_model.fit(x,y)
```

```python
rf_model_pred=rf_model.predict([[6.5]])
print(rf_model_pred)    #This providde random values at differnet execution
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model=RandomForestRegressor(n_estimators=27,random_state=0)
rf_model.fit(x,y)
```

```python
rf_model_pred=rf_model.predict([[6.5]])
print(rf_model_pred)    #Here we get constan values after different execution cuz random_state=0
```
